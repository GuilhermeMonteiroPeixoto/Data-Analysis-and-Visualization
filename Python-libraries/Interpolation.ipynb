# Interpolation - create a function based on fixed data points

```python
import matplotlib.pyplot as plt

a = np.linspace(1, 10, num=20, endpoint=True)
x = a**2
y = np.tan(-a**2/8.0)

plt.plot(x, y, 'o-')
plt.legend(['Dados'], loc='best')
plt.show()
```

```python
from scipy.interpolate import interp1d

f = interp1d(x, y, kind='cubic')
xnew = np.linspace(1, 100, num=100, endpoint=True)

plt.plot(xnew,f(xnew),'r-')
plt.plot(x, y, 'o-')
plt.legend(['Dados','Interp polinomial'], loc='best')
plt.show()
```
