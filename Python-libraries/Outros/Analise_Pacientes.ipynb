```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')
data.drop(['time', 'DEATH_EVENT'], axis=1,inplace=True)
```

```python
fumantes = data[data['smoking']==1]
nao_fumantes = data[data['smoking']==0]
diabeticos = data[data['diabetes']==1]
nao_diabeticos = data[data['diabetes']==0]
```

```python
Aux1 = fumantes['anaemia'].value_counts()
Aux2 = fumantes['diabetes'].value_counts()
TA, NTA =  Aux1[1],Aux1[0]
TD, NTD =  Aux2[1],Aux2[0]
print('GRUPO DE FUMANTES')
print('Tem Anemia      :', TA)
print('Não tem Anemia  :', NTA)
print('Tem Diabetes    :', TD)
print('Não tem Diabetes:', NTD)
```

```python
plt.subplot(1,2,1)
labels = 'Com Anemia','Sem Anemia'
plt.pie([TA,NTA],(0.0, 0.05),labels=labels,autopct='%1.1f%%')
plt.title('Gráfico de Anemia')
plt.subplot(1,2,2)
labels2 = 'Com Diabetes','Sem Diabetes'
plt.pie([TD,NTD],(0.0, 0.05),labels=labels2,autopct='%1.1f%%')
plt.title('Gráfico de Diabetes')
plt.show()
```
